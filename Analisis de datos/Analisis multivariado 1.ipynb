```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
```

```python
df = sns.load_dataset('iris')
df.iloc[:,:-1]
```

```python
pca_model = PCA(n_components= 2)
df_pca = pca_model.fit_transform(df.iloc[:,:-1])
df_pca
```

```python
plt.scatter(df_pca[:,0], df_pca[:,1], c = df['species'].astype('category').cat.codes)
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title('Analisis de componentes principales')
plt.show()
```

```python
sns.displot(data=df, x = 'sepal_length', hue= 'species', kde= True)
```

```python
df['sepal_length_log']= np.log10(df['sepal_length'])
sns.displot(data=df, x = 'sepal_length_log', hue= 'species', kde= True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
#estandarizacion
std_scalater = StandardScaler()
df_std_scalater = std_scalater.fit_transform(df.iloc[:,:-2])
df_std_scalater
#a cuantas desviaciones estandar esta de la media
```

```python
#normalizacion
mm_scalater= MinMaxScaler()
df_normalizate = mm_scalater.fit_transform(df.iloc[:,:-2])
df_normalizate
```

```python
#Correlacion de pearson
from scipy.stats import pearsonr
```

```python
correlacion, _ = pearsonr(df['sepal_length'], df['petal_length'])
print(f'La correlacion entre sepal length y petal length es: {correlacion}')
```

```python
#analisis de clusters
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3, random_state=42)
df['cluster']= model.fit_predict(df.iloc[:,:-2])
df
```

```python
sns.scatterplot(data =df, x ='sepal_length', y='petal_length', hue='cluster')
plt.show()
```

```python
#calcular el z-score
df['z_score_sepal_length']= np.abs((df['sepal_length']-df['sepal_length'].mean())/df['sepal_length'].std())
df
```

```python
outliers = df[df['z_score_sepal_length']>2]
outliers
```

```python
sns.scatterplot(data =df, x ='sepal_length', y='petal_length', hue='species')
sns.scatterplot(data =outliers, x ='sepal_length', y='petal_length', c='r')
plt.show()
```
