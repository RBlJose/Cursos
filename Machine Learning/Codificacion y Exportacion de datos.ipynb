```python
import numpy as np
import matplotlib.pyplot as plt

data= np.random.rand(1000)*100-50  
plt.plot(data)
plt.show()
```

```python
def min_max_scaler(data):
    data_min = np.min(data)
    data_max = np.max(data)
    return (data - data_min) / (data_max - data_min)
```

```python
normal_data = min_max_scaler(data)
plt.plot(normal_data)
plt.show()
```

```python
def standar_sale(data):
    mean= np.mean(data)
    std= np.std(data)
    return(data-mean)/std
```

```python
std_data = standar_sale(data)
plt.plot(std_data)
plt.show()
```
