```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR     
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
boston = pd.read_csv('boston.csv', header=1)
boston
```

```python
x= boston.iloc[:, 5:6]
y= boston.iloc[:, 13:14]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2)
```

```python
arb_desc= DecisionTreeRegressor(max_depth= 5)
```

```python
arb_desc.fit(x_train, y_train)
```

```python
y_pred= arb_desc.predict(x_test)
y_pred, y_test
```

```python
df_pred= pd.DataFrame(y_pred, columns=['PREDICCION'])

# Concatena los DataFrames x y df_pred
conjunto_x = pd.concat([x_test, df_pred], axis=1)
```

```python
x_teste= pd.to_numeric(conjunto_x['RM'])
x_teste
xsd= conjunto_x.to_numpy
xsd
```

```python
x_grid= np.arange(min(x_teste), max(x_teste), 0.1)
x_grid= x_grid.reshape(len(x_grid), 1)
x_grid
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_grid, arb_desc.predict(x_grid), c= 'r', linewidth= 3)
plt.show()
```

```python
arb_desc.score(x_train, y_train)
```
