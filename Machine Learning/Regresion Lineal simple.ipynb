```python
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
boston = pd.read_csv('boston.csv', header=1)
boston
```

```python
print(boston.describe())
```

```python
boston.info()
```

```python
x = pd.DataFrame(boston['RM'])
y = pd.DataFrame(boston['MEDV'])
```

```python
plt.scatter(x, y , c= 'b')
plt.show
```

```python
#Separamos el modelo en entrenamiento y prueba para el algoritmo
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2)
```

```python
#Definimos el modelo a entrenar
lineal_regression= linear_model.LinearRegression()
```

```python
#Entrenamos el modelo
lineal_regression.fit(x_train, y_train)
```

```python
#Realizamos la prediccion del modelo
pred= lineal_regression.predict(x_test)

df_pred = pd.DataFrame(pred, columns=['PREDICCION'])

# Concatena los DataFrames x y df_pred
conjunto_x = pd.concat([y, df_pred], axis=1)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Imprime el conjunto resultante
print(conjunto_x)
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, pred, color= 'r', linewidth= 3)
plt.show()
```

```python
#valore del coeficiente
lineal_regression.coef_
#valor del intercepto
lineal_regression.intercept_
print(f'y= {lineal_regression.coef_}x + {lineal_regression.intercept_}')
```

```python
#precision 
lineal_regression.score(x_train, y_train)
#entre mas se acerca a cero peor es el modelo y si se acerca a 1 su ajuste es preciso
```
