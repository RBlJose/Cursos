```python
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
data = pd.read_csv('salary_Data.csv')
data
```

```python
#Separar las caracteristicas y el target
x= data[['YearsExperience']]
y= data[['Salary']]
```

```python
#Dividir por prueba y entrenamiento
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2)
```

```python
x_test= pd.DataFrame(
    {
    'YearsExperience': [2.3]
}
)
```

```python
#Crear el modelo de regresion lineal
model= LinearRegression()
model.fit(x_train, y_train)
```

```python
#Hacer predicciones sobre el conjunto de prueba
y_pred = model.predict(x_test)
y_pred
```

```python
#Calcular metricas de rendimineto
EC= ((y_test - y_pred)**2).sum()
EC[0]
```

```python
#Error cuadratico medio
ECM = EC/len(y_test)
ECM[0]
```

```python
#Raiz del error cuadratico medio
RMSE = ECM**0.5
RMSE[0]
```

```python
#Error absoluto medio
EAM = abs(y_test - y_pred).mean()
EAM[0]
```

```python
#coeficiente de determinacion
residuos = (y_test - y_pred)
explained_variance = residuos.var()
total_variance = y_test.var() 
R2 = 1-(explained_variance/total_variance)
R2
```
